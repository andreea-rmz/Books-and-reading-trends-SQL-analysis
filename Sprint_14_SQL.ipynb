## 2. CONSULTAS SQL

## 2.1 Número de libros publicados después del 1 de enero de 2000

```python


query = """
SELECT * 
FROM books
WHERE publication_date > '2000-01-01'
ORDER BY publication_date ASC;
"""
df_books = pd.io.sql.read_sql(query, con = engine)
display(df_books)
```

Existen 819 libros publicados a partir del 2000. 

## 2.2 Número de reseñas de usuarios y calificación promedio para cada libro

```python
query = """

SELECT books.book_id,
       books.title,
       COUNT(DISTINCT reviews.review_id) AS review_count,
       ROUND(AVG(ratings.rating), 2)    AS avg_rating
FROM books
LEFT JOIN reviews ON books.book_id = reviews.book_id
LEFT JOIN ratings ON books.book_id = ratings.book_id
WHERE books.publication_date > DATE '2000-01-01'
GROUP BY books.book_id, books.title
ORDER BY avg_rating DESC, review_count DESC;

"""
df_rate_review = pd.io.sql.read_sql(query, con = engine)
display(df_rate_review)
```

Entre los 819 libros, algunos alcanzaron calificaciones promedio altas (5.00) con varias reseñas, como A Dirty Job, Moneyball o School's Out—Forever. También se detectan títulos con calificaciones bajas (por ejemplo Drowning Ruth con 2.00). Todos estos son libros publicados a partir del 2000.

## 2.3 Identifica la editorial que ha publicado el mayor número de libros con más de 50 páginas (esto te ayudará a excluir folletos y publicaciones similares de tu análisis)

```python
query = """


SELECT publishers.publisher_id,
       publishers.publisher,
       COUNT(books.book_id) AS total_books_over_50
FROM books
JOIN publishers ON books.publisher_id = publishers.publisher_id
WHERE books.num_pages > 50
    AND books.publication_date > DATE '2000-01-01'
GROUP BY publishers.publisher_id, publishers.publisher
ORDER BY total_books_over_50 DESC
LIMIT 20;

"""

over_fifty = pd.io.sql.read_sql(query, con = engine)
display(over_fifty)

```

La editorial penguin Books tiene la mayor cantidad de libros con más de 50 páginas (y publicados a partir del año 2000), con un total de 35 libros.

## 2.4 Identifica al autor que tiene la más alta calificación promedio del libro: mira solo los libros con al menos 50 calificaciones.

```python
query= """
SELECT authors.author_id,
       authors.author,
       ROUND(AVG(ratings.rating), 2) AS avg_rating,
       COUNT(ratings.rating) AS total_ratings
FROM books
JOIN authors ON books.author_id = authors.author_id
JOIN ratings ON books.book_id = ratings.book_id
WHERE books.publication_date > DATE '2000-01-01'
GROUP BY authors.author_id, authors.author
HAVING COUNT(ratings.rating) >= 50
ORDER BY total_ratings DESC
LIMIT 20;
"""
fifty_ratings = pd.io.sql.read_sql(query, con = engine) 
display(fifty_ratings)
```

El autor con calificaciones más altas promedio y al menos 50 calificaciones es Diana Gabaldon con una calificación de 4.3 y 50 calificaciones. Es importante señalar que la base de datos presenta registros con hasta 232 calificaciones. Sin embargo, esta incluye a múltiples autores, como es el caso de J.K. Rowling/Mary GrandPré con 232 calificaciones y un rating de 4.29

## 2.5 Encuentra el número promedio de reseñas de texto entre los usuarios que calificaron más de 50 libros.

```python

query = """
SELECT ROUND(AVG(user_reviews.total_reviews), 2) AS avg_reviews_among_active_users
FROM (
    SELECT ratings.username,
           COUNT(DISTINCT reviews.review_id) AS total_reviews
    FROM ratings
    LEFT JOIN reviews ON ratings.username = reviews.username
    GROUP BY ratings.username
    HAVING COUNT(ratings.rating_id) > 50
) AS user_reviews;


"""
active_users = pd.io.sql.read_sql(query, con = engine) 
display(active_users)

```

En el caso de usuarios que han calificado más de 50 libros, el promedio de reseñas es de 17 reseñas por usuario.
